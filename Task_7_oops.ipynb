```python
# problem 1: smart light controller

# create a class for smart light
class smart_light:
    # constructor to initialize light name and status
    def __init__(self, name):
        self.name = name
        self.status = "off"  # default status is off
    
    # method to turn on the light
    def turn_on(self):
        self.status = "on"
    
    # method to turn off the light
    def turn_off(self):
        self.status = "off"
    
    # method to display current status
    def display_status(self):
        print(f"{self.name} is {self.status.upper()}")

# sample input
bedroom_light = smart_light("bedroom light")
bedroom_light.turn_on()
bedroom_light.display_status()  # output: bedroom light is ON
```

```python
# problem 2: employee id card system

# create a class for employee
class employee:
    # constructor to initialize employee name, id, and department
    def __init__(self, name, emp_id, dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept
    
    # method to display id card details
    def display_id_card(self):
        print("employee id card")
        print(f"name: {self.name}")
        print(f"id: {self.emp_id}")
        print(f"department: {self.dept}")

# sample input
rahul = employee("rahul", "emp102", "it")
rahul.display_id_card()
```

```python
# problem 3: mobile contact record

# create a class for contact
class contact:
    # constructor to initialize contact name and phone number
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    
    # method to display contact details
    def display_contact(self):
        print("contact saved")
        print(f"name: {self.name}")
        print(f"phone: {self.phone}")

# sample input
anita = contact("anita", 9876543210)
anita.display_contact()
```

```python
# problem 4: product price tag generator

# create a class for product
class product:
    # constructor to initialize product name and price
    def __init__(self, name, price):
        self.name = name
        self.price = price
    
    # method to display formatted price tag
    def display_price_tag(self):
        print(f"product: {self.name}")
        print(f"price: ₹{self.price}")

# sample input
headphones = product("headphones", 2499)
headphones.display_price_tag()
```

```python
# problem 5: movie rating display system

# create a class for movie
class movie:
    # constructor to initialize movie name and rating
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    
    # method to display movie details
    def display_movie(self):
        print(f"movie: {self.name}")
        print(f"rating: {self.rating} / 5")

# sample input
inception = movie("inception", 4.8)
inception.display_movie()
```

```python
# problem 6: delivery address manager

# create a class for delivery
class delivery:
    # constructor to initialize customer name and address
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address
    
    # method to display delivery details
    def display_delivery(self):
        print("delivery details")
        print(f"customer: {self.customer}")
        print(f"address: {self.address}")

# sample input
suman = delivery("suman", "hyderabad")
suman.display_delivery()
```
