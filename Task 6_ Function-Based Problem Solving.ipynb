```python
def parking_system(capacity, logs):
    current = 0          # current parked vehicles
    peak = 0             # peak parking usage

    for log in logs:     # loop through vehicle logs
        if log == "IN":
            current += 1     # increase count for entry
        elif log == "OUT" and current > 0:
            current -= 1     # decrease count for exit
        
        if current > peak:
            peak = current   # update peak usage

    print("currently parked vehicles:", current)

    if current > capacity:
        print("parking status: full - capacity exceeded")
    elif current == capacity:
        print("parking status: full")  
    else:
        print("parking status: available")

    print("peak usage:", peak)


# sample input
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

parking_system(capacity, logs)
```

```python
def delivery_time(distance, traffic, weather):
    base_time = distance * 5    # assume 5 minutes per km

    # traffic adjustment
    if traffic == "High":
        base_time += 15
    elif traffic == "Medium":
        base_time += 10
    else:
        base_time += 5

    # weather adjustment
    if weather == "Rainy":
        base_time += 10
    elif weather == "Storm":
        base_time += 20

    print("estimated delivery time:", base_time, "minutes")


# sample input
delivery_time(8, "High", "Rainy")
```

```python
def theatre_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)      # count booked seats
    occupancy = (booked_count / total_seats) * 100

    print("occupancy:", round(occupancy), "%")

    if occupancy == 100:
        print("show status: housefull")
    elif occupancy >= 75:
        print("show status: almost full")
    else:
        print("show status: seats available")


# sample input
total = 200
booked = [1] * 150

theatre_occupancy(total, booked)
```

```python
def server_load(cpu_readings):
    total = 0

    for reading in cpu_readings:   # loop to calculate total
        total += reading

    average = total / len(cpu_readings)

    print("average cpu load:", round(average), "%")

    if average < 50:
        print("server status: normal")
    elif 50 <= average <= 80:
        print("server status: warning")
    else:
        print("server status: critical")


# sample input
cpu = [45, 60, 70, 85, 90]
server_load(cpu)
```

```python
def classroom_monitor(resources):
    overused = []    # list to store overused resources

    for resource, hours in resources.items():
        if hours > 8:      # assume more than 8 hours is overuse
            overused.append(resource)

    if overused:
        print("overused resources:", ", ".join(overused))
        print("energy alert: yes")
    else:
        print("overused resources: none")
        print("energy alert: no")


# sample input
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_monitor(usage)
```

```python
def event_registration(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "closed"
    else:
        confirmed = registrations
        waitlist = 0
        status = "open"

    print("confirmed registrations:", confirmed)
    print("waitlisted users:", waitlist)
    print("registration status:", status)


# sample input
event_registration(100, 105)
```
