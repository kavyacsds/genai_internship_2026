1. User Login Check
Problem Statement:
Given a username and password, check whether login is successful.
username = "admin"
password = "1234"


```python
# given correct credentials (stored in system)
correct_username = "admin"
correct_password = "1234"

# Taking input from user
username = input("Enter username: ")
password = input("Enter password: ")

# Checking whether entered credentials match
if username == correct_username and password == correct_password:
    print("login successful")
else:
    print("invalid Crdentials")
```

2. Pass / Fail Analyzer
Problem Statement:
Analyze student results.
marks = [45, 78, 90, 33, 60]


```python
# Given list of student marks
marks = [45, 78, 90, 33, 60]

# Initialize counters
pass_c = 0
fail_c = 0

# Loop through each student's marks
for mark in marks:
    
    # Check passing condition
    if mark >= 50:
        pass_c += 1   # pass counter
    else:
        fail_c+= 1   # fail counter

# Display final result
print("total student :", len(marks))
print("total Pass Students :", pass_c)
print("total fail Students :", fail_c)
```

3. Simple Data Cleaner
Problem Statement:
Clean and standardize user names.
names = [" Alice ", "bob", " CHARLIE "]


```python
# Given list of names
names = [" Alice ", "bob", " CHARLIE "]

# Empty list to store cleaned names
clean_names = []

# Loop through each name
for name in names:
    
    # Step 1: Remove extra spaces using strip()
    name = name.strip()
    
    # Step 2: Convert to lowercase
    name = name.lower()
    
    # Add cleaned name to new list
    clean_names.append(name)

#  cleaned list
print("clened Names:", clean_names)
```

4. Message Length Analyzer
Problem Statement:
Analyze message sizes.
messages = ["Hi", "Welcome to the platform", "OK"]


```python
messages = ["Hi", "Welcome to the platform", "OK"]

# Loop through each message
for message in messages:
    
    # Calculate length of message
    length = len(message)
    
    print("message:", message)
    print("length:", length)
    
    if length > 10:
        print("status: long Message")
    else:
        print("status: short Message")
    
```
