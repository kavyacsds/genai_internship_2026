Problem 1: Mobile Recharge Validation System

```python
# function to validate mobile recharge
def recharge_validation():

    # list of valid recharge plans
    valid_plans = [199, 299, 399, 599]

    while True:
        # taking recharge amount from user
        amount = int(input("enter recharge amount: "))

        # check if amount is less than 50
        if amount < 50:
            print("recharge amount must be at least 50 rupees")
            print("please try again\n")
            continue   # allow user to retry

        # check if amount is in valid plans
        if amount in valid_plans:
            print("recharge successful")
            break   # exit loop if valid
        else:
            print("invalid plan selected")
            print("available plans are:", valid_plans)
            print("please try again\n")


# calling the function
recharge_validation()
```

Problem 4: Cab Fare Estimator with Retry Option


```python
# function to calculate cab fare
def cab_fare_calculator():

    while True:

        # taking distance input from user
        distance = float(input("enter distance in km: "))

        # asking if it is peak hour
        peak = input("is it peak hour? (yes/no): ").lower()

        # base fare
        base_fare = 50

        # cost per km
        per_km_rate = 12

        # calculating normal fare
        total_fare = base_fare + (distance * per_km_rate)

        # if peak hour add 25 percent extra
        if peak == "yes":
            total_fare = total_fare + (0.25 * total_fare)

        # printing total fare
        print("total fare is:", total_fare)

        # asking user if they want to retry
        retry = input("do you want to calculate again? (yes/no): ").lower()

        if retry != "yes":
            print("thank you for using cab fare calculator")
            break


# calling the function
cab_fare_calculator()

        
```

Problem 5: Employee Attendance Eligibility Checker

```python
# function to check employee attendance eligibility
def attendance_checker(attendance_list):

    # total number of working days
    total_days = len(attendance_list)

    # count number of present days
    present_days = attendance_list.count("P")

    # calculate attendance percentage
    percentage = (present_days / total_days) * 100

    # print attendance percentage
    print("attendance percentage:", percentage)

    # check eligibility condition
    if percentage >= 75:
        return "eligible"
    else:
        return "not eligible"


# example attendance list
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]

# calling function
result = attendance_checker(attendance)
print(result)
```

Problem 6: Password Strength Checker
Problem Statement


```python
# function to check password strength
def password_strength_checker(password):

    # check minimum length
    if len(password) < 8:
        return "weak password - minimum length should be 8"

    # variables to track conditions
    has_digit = False
    has_special = False

    # loop through each character in password
    for ch in password:

        # check if character is digit
        if ch.isdigit():
            has_digit = True

        # check if character is special character
        if ch in "@#$":
            has_special = True

    # final validation
    if has_digit and has_special:
        return "strong password"
    else:
        return "weak password - must contain at least one digit and one special character"


# example usage
pwd = input("enter password: ")
result = password_strength_checker(pwd)
print(result)
```
