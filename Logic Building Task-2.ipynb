Problem 1: Unique Words in a Sentence
Problem Statement
Given a sentence, identify all unique words and display:
The count of unique words
The unique words themselves


```python
inp="python is easy and python is powerful"
splitt=inp.split(' ')
list_of_words=list(splitt)
unique_words=set(list_of_words)
length_of_words=len(unique_words)
print(length_of_words)
print(unique_words)
```

Problem 2: Highest Salary from Employee Data
Problem Statement
Store employee names and their salaries, and determine which employee has the highest salary.



```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_employee = ""
highest_salary = 0

for name, salary in employees.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_employee = name

print("Highest Salary:", highest_employee, "-", highest_salary)
```

Problem 3: Find Maximum and Minimum Values
Problem Statement
Find the maximum and minimum values from a list of numbers.



```python
# Given list
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

# Traverse the list
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Output
print("list:", numbers)
print("max:", maximum)
print("min:", minimum)
```

Problem 4: Count Products Above a Price Threshold
Problem Statement
Count how many products have a price greater than 1000.


```python
# given list
prices = [450, 1200, 899, 1500, 300]

count = 0

# checking each price
for price in prices:
    if price > 1000:
        count += 1

# output
print("products above 1000:", count)
```

Problem 5: Calculate Attendance Percentage Problem Statement Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).

```python
# given list
attendance = ["P", "P", "A", "P", "P"]

present_count = 0

# count Present days
for day in attendance:
    if day == "P":
        present_count += 1

# total days
total_days = len(attendance)

# percentage calculation
percentage = (present_count / total_days) * 100

# output
print("attendance prcentage:", percentage)
```

Problem 7: Count Character Frequency Problem Statement Count how many times each character appears in a given string.

```python
# given string
text = "pythonp"

frequency = {}

# count characters
for ch in text:
    if ch in frequency:
        frequency[ch] += 1
    else:
        frequency[ch] = 1

# output
print(frequency)
```

Problem 10: Calculate Average Marks
Problem Statement
Calculate the average marks of a student from a list of marks.


```python
# Given marks
marks = [70, 75, 80, 65]

# Calculate average
total = sum(marks)
average = total / len(marks)

print("Average Marks:", average)
```
