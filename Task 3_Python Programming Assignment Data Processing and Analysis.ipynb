Problem Statement 1: Employee Performance Bonus Eligibility


```python
# problem 1: employee performance bonus eligibility

# dictionary storing employee names and their performance scores
employees = {
    "ravi": 92,
    "anita": 88,
    "kiran": 92,
    "suresh": 85
}

# find the highest performance score
highest_score = max(employees.values())

# find all employees who have the highest score (handle tie case)
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

# display result
print("top performers eligible for bonus:", ", ".join(top_performers), "(score:", highest_score, ")")
```

Problem Statement 2: Search Query Keyword Analysis


```python
# problem 2: search query keyword analysis

# input search query
query = "buy mobile phone buy phone online"

# convert to lowercase
query = query.lower()

# remove punctuation (if any)
# here we manually remove common punctuation
for char in ".,!?":
    query = query.replace(char, "")

# split sentence into words
words = query.split()

# dictionary to store word frequency
word_count = {}

# count frequency of each word
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# dictionary to store only words appearing more than once
result = {}

for word, count in word_count.items():
    if count > 1:
        result[word] = count

# display result
print(result)
```

Problem Statement 3: Sensor Data Validation

```python
# problem 3: sensor data validation

# list of sensor readings
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# list to store valid readings as (hour_index, reading_value)
valid_readings = []

# iterate through list using index
for index in range(len(sensor_readings)):
    value = sensor_readings[index]
    
    # check if reading is even (valid)
    if value % 2 == 0:
        valid_readings.append((index, value))

# display result
print("valid sensor readings (hour, value):")
print(valid_readings)
```

Problem Statement 4: Email Domain Usage Analysis

```python
# problem 4: email domain usage analysis

# list of emails
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# dictionary to store domain count
domain_count = {}

# total number of emails
total = len(emails)

# extract domain and count
for email in emails:
    # split email using '@'
    domain = email.split("@")[1]
    
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# calculate percentage usage
for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(domain + ":", str(int(percentage)) + "%")
```

Problem Statement 5: Sales Spike Detection

```python
# problem 5: sales spike detection

# list of daily sales
sales = [1200, 1500, 900, 2200, 1400, 3000]

# calculate average sales
average_sales = sum(sales) / len(sales)

# threshold is 30 percent above average
threshold = average_sales * 1.3

# check each day for spike
for day in range(len(sales)):
    if sales[day] > threshold:
        print("day", day + 1, ":", sales[day])
```

Problem Statement 6: Duplicate User ID Detection

```python
# problem 6: duplicate user id detection

# list of user ids
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# dictionary to count occurrences
id_count = {}

# count each user id
for user in user_ids:
    if user in id_count:
        id_count[user] += 1
    else:
        id_count[user] = 1

# display only duplicates
for user, count in id_count.items():
    if count > 1:
        print(user, "→", count, "times")
```
